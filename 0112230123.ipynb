#answer to the question number 2:

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('vgsales.csv')


avg_glob_sale = df['Global_Sales'].mean()


filter_df = df[df['Global_Sales'] >= avg_glob_sale].copy()


avg_eu_sales = filter_df['EU_Sales'].mean()

```

```python
print("\n")
print("average eu Sales:", avg_eu_sales)

```

```python
scaler = MinMaxScaler()


filter_df['Global_Sales_Normalized'] = scaler.fit_transform(
    filter_df[['Global_Sales']]
)
```

```python
print(filter_df.head())
```
